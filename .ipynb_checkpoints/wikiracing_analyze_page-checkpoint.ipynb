```python
import requests
from bs4 import BeautifulSoup, SoupStrainer
import psycopg2 as pg
```

```python
main_body = r'https://uk.wikipedia.org/wiki/'
r = requests.get(main_body+'Дружба')
```

```python
bs = BeautifulSoup(r.content, 'lxml')
for link in bs.find_all('a', href=True):
    if link['href'].startswith('/wiki'):
        print(f'{link.text}: {link["href"]}')
```

```python
import requests
main_body = 'https://uk.wikipedia.org/wiki/'
r = requests.get(main_body+'Аденозинтрифосфат')
```

```python
import pandas as pd 

bs = BeautifulSoup(r.content, 'lxml')
d = {'link': [], 'class': [], 'text': [], 'title': []}

for link in bs.find('div', attrs={'class': 'mw-parser-output'}).find_all('a', href=True):
    d['link'] += [link['href']]
    d['class'] +=  [link.get('class')]
    d['text'] += [link.text]
    d['title'] += [link.get("title")]
df = pd.DataFrame(d)
df
df['class'] = df['class'].apply(lambda x: x if x is None else ' '.join(x))
print(df['class'].unique())
# df[(df['class'].isna()) & (df['link'].str.startswith('/wiki/'))]
# 'interlanguage-link-target'
df[(df.text == 'Вітамін К') & (df['class'].isin(['mw-redirect', None]) & df['link'].str.startswith('/wiki')) & df['title'].notna()]
# df[df['title'].notna() & (df['class'].isin(['mw-redirect', None])) & df['link'].str.startswith('/wiki')  \
#    & (df.text == 'Вітамін K')]

# ['mw-redirect', None, 'internal']
```
